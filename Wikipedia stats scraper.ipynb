# Wikipedia Stats Scraper 

This is a scraper for the statistics from [this kind of page](https://xtools.wmflabs.org/articleinfo/en.wikipedia.org/Area%20chart).

We need:

* Page size
* Pageviews (from the beginning)
* Links to this page
* Prose (everything)

```python
# libraries for requesting url and scraping jsons
import requests
```

```python
url = 'https://xtools.wmflabs.org/articleinfo/en.wikipedia.org/Area%20chart'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36'}
r = requests.get(url, headers=headers)

```

```python
print(r)
```

```python
html = r.content.decode()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
text = []
for link in soup.find_all(class_="stat-list"):
    text.append(link.getText())
```

```python
print(text)
```

```python
str1 = ''.join(text)
```

```python
print(str1)
```

```python
import re
```

```python
page_size = re.findall('Page size\n.+ bytes', str1)
links_number = re.findall('Links to this page\n\n.+', str1)
prose = re.findall('Characters\n.+\n\n\nWords\n.+\n\n\nSections\n.+\n\n\nReferences\n.+\n\n\nUnique references\n.+', str1)
```

```python
print(page_size)
print(links_number)
print(prose)
```
