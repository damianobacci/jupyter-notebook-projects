```python
base_url = "https://www.privateislandsonline.com/search?region=&diversion=&availability=sale&price_range=0%3A50000000&size_range=0%3A1000&q=&view%5B0%5D=1&order=created_at%3ADESC&page=74"

# ci sono isole fino a 75 pagine (9 isole per pagina)
```

```python
# libraries for requesting url and scraping jsons
import requests
import json
```

```python
r = requests.get(base_url)
```

```python
from bs4 import BeautifulSoup
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.select('div[class="grid-content island-content"] span[class="name"]')
```

```python
names = [r.text for r in results]
```

```python
names
```

```python
dimensions = soup.select('div[class="grid-content island-content"] span[class="num"]')
```

```python
acres = [r.text for r in dimensions]
```

```python
acres
```

```python
len(acres)
```

```python
prices = soup.select('div[class="grid-content island-content"] div[class="serv-price-wrapper"]')
```

```python
dollars = [r.text for r in prices]
        
```

```python
dollars
```

```python
len(dollars)
```

```python
locations = soup.select('div[class="grid-content island-content"] div[class="list-name"]')
```

```python
places = [r.text for r in locations]
```

```python
places
```

```python
len(places)
```

```python
list(zip(names, places, dollars))
```
