### Initial data cleaning

We are going to read the "Not Stated" values as NaN to uniform all values, and drop the columns that we don't need for the analysis.

```python
dete_survey = pd.read_csv('/Users/bacci/Datasets/dete_survey.csv', na_values='Not Stated') #reading Not Stated as NaN
```

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis=1) #dropping columns that are not useful for the analysis
```

```python
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis=1) #dropping columns that are not useful for the analysis
```

We need to uniform the names of the relevant columns in both the Dete and the Tafe surveys, to better filter the data with columns that are common for both datasets.

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.replace(" ", "_").str.strip().str.lower()
```

```python
dete_survey_updated.columns
```

```python
mapping = {'Record ID': 'id', 'CESSATION YEAR': 'cease_date', 'Reason for ceasing employment': 'separationtype', 'Gender. What is your Gender?': 'gender', 'CurrentAge. Current Age': 'age',
       'Employment Type. Employment Type': 'employment_status',
       'Classification. Classification': 'position',
       'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
       'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}
```

```python
tafe_survey_updated = tafe_survey_updated.rename(mapping, axis=1)
```

```python
tafe_survey_updated.columns
```

### Filtering the data for Resignation

Recall that we are looking for unique values in the the `separationtype` column

```python
dete_survey_updated['separationtype'].value_counts()
```

```python
tafe_survey_updated['separationtype'].value_counts()
```

```python
tafe_resignations = tafe_survey_updated[tafe_survey_updated['separationtype'] == 'Resignation'].copy()
```

```python
dete_survey_updated['separationtype'] = dete_survey_updated['separationtype'].str.split('-').str[0]
```

```python
dete_resignations = dete_survey_updated[dete_survey_updated['separationtype'] == 'Resignation'].copy()
```

With these steps we have isolated all the relevant rows of the *Resignation* type. We will now proceed to clean the date columns.

```python
dete_resignations['cease_date'].value_counts()
```

In the date resignation we need to extract the year.

```python
dete_resignations['cease_date'] = dete_resignations['cease_date'].str.split('/').str[-1]
```

```python
dete_resignations['cease_date'] = dete_resignations['cease_date'].astype('float')
```

```python
dete_resignations['dete_start_date'].value_counts()
```

```python
tafe_resignations['cease_date'].value_counts()
```

Some of the dates don't really align, but the difference is not very big so we can leave the 2009 years.

We are going to create a column in the dete dataframe that reflect the amount of years in service.

```python
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
dete_resignations['institute_service'].head()
```
